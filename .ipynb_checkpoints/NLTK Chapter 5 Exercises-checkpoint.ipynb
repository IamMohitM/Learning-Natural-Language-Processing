##### 12. Train a bigram tagger with no backoff tagger, and run it on some of the training data. Next, run it on some new data. What happens to the performance of the tagger? Why?

```python
bigram = nltk.BigramTagger(train_sents)
bigram.evaluate(train_sents)
```

```python
bigram.evaluate(test_sents)
```

The performance goes down. A bigram tagger looks at the n-1 tag word n + the word n and assigns the tag which most frequently occurs after n-1 word's tag.  And so in case it has never seen a word, it can not assign a tag to it. Plus if it has seen a word but never after a particular tag it will assign None too.

```python
cfd = nltk.ConditionalFreqDist((word, tag) for word, tag in brown.tagged_words())
```

```python
cfd
```

```python
total_tags = [(word, cfd[word].B(), cfd[word].keys()) for word in cfd.conditions()]
print(total_tags[:10])
```

```python
sorted(total_tags, key = itemgetter(1), reverse = True)[0:10]
```

```python
max(total_tags, key = itemgetter(1))
```

```python
tags_af_noun = [(t1, t2) for (w1, t1), (w2, t2) in nltk.bigrams(brown_tagged_words) if t1[:2] =='NN']
```

**30. Preprocess the Brown News data by replacing low frequency words with UNK, but leaving the tags untouched. Now train and evaluate a bigram tagger on this data. How much does this help? What is the contribution of the unigram tagger and default tagger now?**

```python
brown_news = brown.tagged_words(categories = 'news')
fd = nltk.FreqDist(w for w, t in brown_news)
most_freq  = fd.most_common(1000)
most_freq
```

```python
for i, w, _ in enumerate(brown_news):
    if w not in most_freq:
        print(w)
        brown_news[i] = ('UNK','_')
```

```python
for w, _ in brown_news:
    if w is 'UNK':
        print(w,_)
```
